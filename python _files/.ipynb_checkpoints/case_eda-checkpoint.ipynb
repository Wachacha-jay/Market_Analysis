## Explorarory Data Analysis

```python
#import the nessesary libraries to used
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
from sklearn.metrics import mean_absolute_error
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
import time
```

```python
#import the csv data file
df = pd.read_csv(r"C:\Users\jay\Desktop\Kwanza Tukule\Data\Cleaned_data.csv")
df.head()
```

```python
#set date as index
df.set_index("DATE", inplace = True)
```

```python
grouped_df = df.groupby(['ANONYMIZED CATEGORY', 'ANONYMIZED BUSINESS']).agg(
    Total_Quantity=('QUANTITY', 'sum'),
    Total_Value=('UNIT PRICE', 'sum')
).reset_index().sort_values(by = "Total_Value", ascending = False).head(10)
```

```python
grouped_df
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=grouped_df, x='ANONYMIZED CATEGORY', y='Total_Quantity', hue='ANONYMIZED BUSINESS')
plt.title('Total Quantity by Anonymized Category')
plt.ylabel('Total Quantity')
plt.xlabel('Anonymized Category')
plt.legend(title='Anonymized Business')
plt.legend(title='Anonymized Business', loc='upper left', bbox_to_anchor=(1, 1), bbox_transform=plt.gca().transAxes)

#plt.savefig('total_quantity_by_category.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
grouped_df.info()
```

```python
grouped_df.pivot(index='ANONYMIZED CATEGORY', columns='ANONYMIZED BUSINESS', values='Total_Quantity').plot(kind='bar', stacked=True)
plt.title('Stacked Bar Chart of Total Quantity by Anonymized Category')
plt.ylabel('Total Quantity')
plt.xlabel('Anonymized Category')
plt.legend(title='Anonymized Business', loc='upper left', bbox_to_anchor=(1, 1), bbox_transform=plt.gca().transAxes)
plt.show()
```

```python
# Create a pivot table for Total Value
pivot_table = grouped_df.pivot(index='ANONYMIZED CATEGORY', columns='ANONYMIZED BUSINESS', values='Total_Value')

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap="YlGnBu")
plt.title('Heatmap of Total Value')

# Save the heatmap as an image
plt.savefig('heatmap_total_value.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
#calculate the total value by multiplying quantities sold and the unit price
df['Total_Value'] = df['QUANTITY'] * df['UNIT PRICE']

# Group by Month_Year and sum Quantity and Total_Value
monthly_sales = df.groupby('Month_Year').agg({'QUANTITY': 'sum', 'Total_Value': 'sum'}).reset_index()

# Plotting
plt.figure(figsize=(14, 7))

# Plot Total Sales Value
plt.subplot(2, 1, 1)
plt.plot(monthly_sales['Month_Year'], monthly_sales['Total_Value'], marker='o', label='Total Value', color='b')
plt.title('Monthly Sales Trends: Value')
plt.ylabel('Sales Value')
plt.xticks(rotation=45)
plt.grid()
plt.legend()

# Plot Total Quantity Sold
plt.subplot(2, 1, 2)
plt.plot(monthly_sales['Month_Year'], monthly_sales['QUANTITY'], marker='o', label='Total Quantity', color='g')
plt.title('Monthly Sales Trends: Quantity')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.grid()
plt.legend()

# Adjust layout and save the figure
plt.tight_layout()
plt.savefig('monthly_sales_trends.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Calculate Total Value for each product
df['Total_Value'] = df['QUANTITY'] * df['UNIT PRICE']

# Group by ANONYMIZED PRODUCT and aggregate Quantity and Total Value
product_summary = df.groupby('ANONYMIZED PRODUCT').agg(
    Total_Quantity=('QUANTITY', 'sum'),
    Total_Value=('Total_Value', 'sum')
).reset_index()

# Identify the top 5 most frequently purchased products (based on Quantity)
top_5_frequent = product_summary.nlargest(5, 'Total_Quantity')
top_5_frequent
```

```python
# Identify the top 5 most valuable products (based on Value)
top_5_valuable = product_summary.nlargest(5, 'Total_Value')
top_5_valuable
```

```python
#reset index
df.reset_index(inplace = True)
```

```python
# Calculate Total Value for each transaction
df['Total_Value'] = df['QUANTITY'] * df['UNIT PRICE']

# Group by ANONYMIZED BUSINESS and aggregate metrics
business_summary = df.groupby('ANONYMIZED BUSINESS').agg(
    Total_Quantity=('QUANTITY', 'sum'),
    Total_Value=('Total_Value', 'sum'),
    Frequency=('DATE', 'count')
).reset_index()

# Classify businesses into High, Medium, Low Value based on Total_Value
value_bins = pd.qcut(business_summary['Total_Value'], 3, labels=['Low Value', 'Medium Value', 'High Value'])
business_summary['Value_Segment'] = value_bins
```

```python
business_summary[business_summary['Value_Segment'] == 'High Value']
```

```python
business_summary[business_summary['Value_Segment'] == 'Low Value']
```

```python
business_summary[business_summary['Value_Segment'] == 'Medium Value']
```

```python
# Calculate correlation
correlation = df['QUANTITY'].corr(df['Total_Value'])
print(f'Correlation between Quantity and Total Value: {correlation}')

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['QUANTITY'], df['Total_Value'], alpha=0.5)
plt.title('Scatter Plot of Quantity vs Total Value')
plt.xlabel('Quantity')
plt.ylabel('Total Value')
plt.grid()
plt.show()
```

The Correlation Cofficient 0.8 is positive and close to 1 which indicates that there is a strong correlation between quantity and the total value as the quantity increases so does the total value.

```python
monthly_sales["Total_Value"] = monthly_sales["Total_Value"].astype(int)
monthly_sales.info()
```

```python
monthly_sales = monthly_sales.to_frame()
```

```python
monthly_sale = monthly_sales.copy()
# Calculate moving average for trend comparison
monthly_sale['Moving_Avg'] = monthly_sale['Total_Value'].rolling(window=3, min_periods=1).mean()

# Compute percentage change (handling division errors)
monthly_sale['Pct_Change'] = monthly_sale['Total_Value'].pct_change().fillna(0)

# Identify significant anomalies (spikes or drops beyond 2 standard deviations)
threshold = monthly_sale['Pct_Change'].std() * 2
monthly_sale['Anomaly'] = monthly_sale['Pct_Change'].apply(lambda x: 'Spike' if x > threshold else ('Drop' if x < -threshold else 'Normal'))

# Extract anomalies for review
anomalies = monthly_sale[monthly_sale['Anomaly'] != 'Normal'].reset_index()

print(monthly_sale.head(20))
print(anomalies)
```
